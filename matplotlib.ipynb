import matplotlib.pyplot as plt

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
print("\n")
print(y)
print("\n")
print(z)
```

Ploting the data

```python
#sine wave

plt.plot(x,y)

plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

**adding tittle x axis & y axis labls**

```python

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin val')
plt.title('sine wave')
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cosine val')
plt.title('cosine wave')
plt.show()
```

```python
x=np.linspace(-10,10,20) #means 20 evenly value between -10 and 10
y=x**2
plt.plot(x,y)
plt.show()
```

```python
#color
plt.plot(x,y,'g+')
plt.show()
```

```python
plt.plot(x,y,'b-')
plt.show()
```

**multiple line in single graph**

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r-')
plt.show()
```

**BAR PLOT**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','Hindi','French','spanish','German']
people=[100,50,78,200,134]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

**PIE CHART**

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','Hindi','French','spanish','German']
people=[100,50,78,200,134]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

**scatter plot**
  


```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

**3D scatter plot**

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Reds')
plt.show()
```
